```python
# This very ugly segment is the result of a moving stimulus!
# I approximate coordinates for the stimulus (a live crab in this case) to demonstrate functionality.
# In practice the package will still be able to tell if an octopus is looking at a static stimulus,
# but a crab tracking network will be needed to generate accurate crab coordinates

x_init = [0] * 900
x1s = [35] * 100
x2s = [61] * 100
x3s = [96] * 100
x4s = [127] * 100
x5s = [162] * 100
x6s = [180] * 100
x_tail = [0] * 939
crab_xs = x_init + x1s + x2s + x3s + x4s + x5s + x6s + x_tail

y_init = [0] * 900
y1s = [170] * 100
y2s = [182] * 100
y3s = [207] * 100
y4s = [230] * 100
y5s = [240] * 100
y6s = [240] * 100
y_tail = [0] * 939
crab_ys = y_init + y1s + y2s + y3s + y4s + y5s + y6s + y_tail

data = {'x':(crab_xs),'y':(crab_ys)}
crab_df = pd.DataFrame(data)
crab_points = package_stim_points(crab_df)
```

```python
visualize_frame(900,pairs,lines,crab_points,vid)
visualize_frame(1100,pairs,lines,crab_points,vid)
visualize_frame(1400,pairs,lines,crab_points,vid)
```
